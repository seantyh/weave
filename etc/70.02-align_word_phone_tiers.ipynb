```python
from praatio import textgrid
```

## Aligning annotation data

```python
tg = textgrid.openTextgrid("../data/CN_GY_001.TextGrid", False)
words = tg.getTier("words")
phones = tg.getTier("phones")
word_idx = 0
phone_entries = []
cur_word = words.entries[0]
for phone_idx, phone_x in enumerate(phones.entries):
    if (phone_x.start >= cur_word.end):
        word_idx += 1
    cur_word = words.entries[word_idx]
    phone_entries.append((
        cur_word.start, cur_word.end, cur_word.label,
        phone_x.start, phone_x.end, phone_x.label))

assert len(phone_entries) == len(phones.entries)
```
